# 7: Numerical Integration [0:53:35]

* For doing a integration scipy have a **quad** function for single integration.
* For duble integration we can use a **dblquad**.
* In integration should specipy a **Limits**
* after Integration the function give **Two Result** in that **value** and **estimated error**

**Explanation**

```python
'''
∫ 3x^2 + 2x + 6 dx

 3*(x^(2+1)/(2+1)) + 2*(x^(1+1)/(1+1)) + 6x

 3* (x^3)/3 + 2 *(x^2)/2 + 6x

 x^3 + x^2 + 6x

 *Now limit 0
 0

 *Now limit 2
 = (2^3) + (2^2) + 6*2
 = 8 + 4 + 12
 = 24
 '''
```

```python
from scipy import integrate

func = lambda x: 3*x**2 + 2*x + 6
integrate.quad(func,0,2)
        # 0 lower limit
        # 2 upper limit
```

# 8: Double Integral[1:04:40]

* In Double Integral we using a **dblquad( )** Function.
* In this function we have to give **Two set of Limits**

**Explanation**

```python
'''
∫ ∫ xy^2 dx dy

∫ (x^(1+1))/(1+1) (y^2) dy

∫ ((x^2) / 2) * (y^2) dy

((x^2)/2) * (y^(2+1))/(2+1)

((x^2)/2) * ((y^3)/3)

((2^2)/2) * ((1^3)/3)

(4/2) * (1/3)

2 * 0.3333

0.6666

'''
```

```python
func = lambda y, x: x*y**2
integrate.dblquad(func,
                  0,2,
                  0,1)
```

# 13: Eigen Value and Eigen Vector[00:10:28]

**scipy.linalg.eig(array)**

**Explanation**

```python
arr = np.array([
    [1, 2, 3],
    [4, 6, 1],
    [5, 1, 2]
])
a = scipy.linalg.det(arr)
print(a)
```

```python
'''
a = [1, 2, 3]
    [4, 6, 1]
    [5, 1, 2]

* Formula
  (A - λI) x = 0

  A = given Matrix
  λ = Constant
  I = identity Matrix
  x = Unknow Matrix

  | A - λI | = 0


  [1, 2, 3]        [1, 0, 0]
          -  λ
  [4, 6, 1]        [0, 1, 0]
  [5, 1, 2]        [0, 0, 1]


  [1-λ       2,    3]
  [4       6-λ,    1]  = 0
  [5,       1,   2-λ]

 Step 2:
 λ^3 - [ x ]λ^2 + [ y ]λ - |A| = 0

 x = Sum of Digonal Element
    in given matrix
    = 1 + 6 + 2
    = 9

 y = Sum of Digonal Diterment

   d1 = (6*2) - (1)
      = 12 - 1 => 11

   d2 = (1*2) - (5*3)
      = 2 - 15 => -13

   d3 = (1*6) - (4*2)
      = 6 - 8 => -2

   y = d1 + d2 + d3
     = 11 - 13 - 2 => -4

 |A| = Determent of A matrix
     = -73

     λ^3 - 9λ^2 - 4λ + 73 = 0
     find the root that is a Eigen Value

  '''
```

```python
from scipy import linalg
arr = np.array([
    [1, 2],
    [4, 6]
])
eig_val, eig_vec = linalg.eig(arr)
print("eigen value :\n", eig_val)
print('-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-')
print("Eigen Vector :\n", eig_vec)
```

# 17.2: Chi-Squared Test[1:02:00]

* Chi-Squared test used for selecting a column or feature selection

```python
import scipy.stats as st
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataset = sns.load_dataset('tips')
dataset
```

```python
# relate sex and smokers
dataset_table = pd.crosstab(dataset['sex'],
                           dataset['smoker'])
print(dataset_table)
```

```python
obserd = dataset_table.values
print(obserd)
```

* In above code we get a **Observe values**

```python
val = st.chi2_contingency(dataset_table)
print(val)
```

```python
Expected_Value = val[3]
print(Expected_Value)
```

* In above code we get a **Expected Value**

```python
chi_squre = sum([(o-e)**2 / e for o,
        e in zip(obserd, Expected_Value)])
print(chi_squre)
```

```python
# Now add two value to single
chi_squre_stat = chi_squre[0] + chi_squre[1]
print(chi_squre_stat)
```

# 17.21: DOF [1;21:16]

```python
dataset_table
```

```python
no_of_row = len(dataset_table.iloc[0:2, 0])
no_of_col = len(dataset_table.iloc[0, 0:2])
dof = (no_of_row - 1)*(no_of_col - 1)
print(dof)
alpha = 0.05
```

# 17.22: For checking Null Hypothesis Using Critical Value [1:33:00]

**scipy.stats.chi2.ppf()**

```python
#First Method
# finding Critical Value
# ppf - percent point function

critical_value = st.chi2.ppf(q=1-alpha, df=dof)
print(critical_value)
```

```python
if chi_squre_stat >= critical_value:
    print("reject ho two term related")
else:
    print("Retain ho not related")
```

# 17.23: For checking Null Hypothesis Using P Value [1:35:40]

**scipy.stats.chi2.cdf()**

```python
# second method
p_value = 1 - st.chi2.cdf(x=chi_squre_stat,
                          df=dof)
print(p_value)
print("sigificance level: ", alpha)
print(dof)
```

```python
if p_value <= alpha:
    print("reject ho two term related")
else:
    print("Retain ho not related")
```
